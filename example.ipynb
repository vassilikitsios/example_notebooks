# Example plot

* $\sin(x)$

```python
x = np.linspace(0,2*np.pi,100)
y = np.sin(x)

fig = plt.figure()
plt.plot(x,y)
plt.xlabel('$x$')
plt.ylabel('$\sin(x)$')

plt.tight_layout()
```
